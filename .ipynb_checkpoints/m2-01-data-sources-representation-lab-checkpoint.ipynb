Step 3: Write a validation function called validate_registrations(entries) that takes Dataset A and returns two lists: valid_entries and invalid_entries. A valid entry must have a non-empty username, an email containing @, and an age that can be converted to a positive integer. Test your function on Dataset A and print the counts.

```python
def validate_registrations(entries):
    valid_entries = []
    invalid_entries = []
    for dct in entries:
        try:
            age_int = int(dct['age'])
            if (
                dct['username'] and              
                '@' in dct['email'] and          
                age_int > 0                      
            ):
                valid_entries.append(dct)
            else:
                invalid_entries.append(dct)
        except ValueError:
            invalid_entries.append(dct)
    return f" Valid entries:{valid_entries}\nInvalid entries: {invalid_entries}"

validate_registrations(df_a)          
    
```

The **Parquet** format preserved data types most accurately.

Typical size relationship:
Parquet < CSV < JSON

### When would you choose each format?
**CSV**
- Best for data sharing
- Human-readable
- Compatible with almost all tools
- Good for small datasets
- Not ideal for preserving data types

**JSON**
- Useful for web applications and APIs
- Handles nested or semi-structured data
- Flexible structure
- Usually larger file size

**Parquet**
- Best for data analytics and big data workflows
- Preserves data types accurately
- Highly compressed and efficient
- Faster reading and querying performance
- Preferred for machine learning and data engineering pipelines

### Task 3: Row-Major vs Column-Major Access Patterns
This task lets you observe the performance differences between row-level and column-level data access empirically.

Step 1: Create a large DataFrame with 200,000 rows and 20 numeric columns (random float values). Name the columns feature_0 through feature_19.

```python
rows = 200_000
cols = 20
data = np.random.rand(rows, cols)
column_names = [f'feature_{i}' for i in range(cols)]
df = pd.DataFrame(data, columns=column_names)
print(df.shape)
df.head()
```

### Step 2: Measure and compare the time for these two operations:

Operation A — Column access: Compute the mean of feature_0 using the pandas column accessor (df["feature_0"].mean()).

Operation B — Row iteration: Loop through the first 10,000 rows using df.iloc[i] and sum the values of feature_0 manually.

Use %%timeit or %time to measure each. Record the results.

```python
%%timeit
df["feature_0"].mean()
```

```python
%%timeit
total = 0
for i in range(10000):
    total += df.iloc[i]["feature_0"]
mean_manual = total / 10000
```

### Step 3: Now convert the DataFrame to a NumPy array and repeat both operations:

Operation C — NumPy column access: Compute the mean of column 0 (arr[:, 0].mean()).

Operation D — NumPy row access: Access the first 10,000 rows sequentially (arr[i] in a loop) and sum column 0.

Record the results.

```python
arr = df.to_numpy()
```

```python
arr.shape
```

```python
%%timeit
arr[:, 0].mean()
```

```python
%%timeit
total = 0
for i in range(10000):
    total += arr[i][0]

mean_manual = total / 10000
```

Why is column access in pandas faster?

Pandas is column-oriented, storing each column in contiguous memory. Operations like df["feature_0"].mean() are vectorized in C, while row iteration (df.iloc[i]) is slow due to Python-level loops and assembling data from multiple columns.

Why does the gap narrow with NumPy?

NumPy arrays are contiguous in memory, so row access is faster than in pandas. Column operations remain fastest due to optimized vectorized computations.

Memory layout insights

Pandas: column-major structure → best for column-wise operations.

NumPy: contiguous multidimensional arrays → efficient for both rows and columns, but vectorized ops still excel.

# Task 4: Text vs Binary Format Deep Dive
This task explores the text-versus-binary trade-off with a focus on practical consequences.

Step 1: Create a DataFrame with 100,000 rows and 5 numeric columns of random float values (use np.random.seed(42) for reproducibility). Add one string column called category with values randomly chosen from ["alpha", "beta", "gamma"].

```python
np.random.seed(42)
numeric_data = np.random.rand(100000, 5)
categories = np.random.choice(["alpha", "beta", "gamma"], size=100000)
df2 = pd.DataFrame(numeric_data, columns=[f"feature_{i}" for i in range(5)])
df2['category'] = categories
print(df2.head())
```

```python
csv_file = "data.csv"
df2.to_csv(csv_file, index=False)
parquet_file = "data.parquet"
df2.to_parquet(parquet_file, index=False)
csv_size = os.path.getsize(csv_file) / (1024*1024)       
parquet_size = os.path.getsize(parquet_file) / (1024*1024)  
print(f"CSV file size: {csv_size:.2f} MB")
print(f"Parquet file size: {parquet_size:.2f} MB")
```

```python
df_csv = pd.read_csv("data.csv")
print("CSV dtypes:\n", df_csv.dtypes)

# Read Parquet
df_parquet = pd.read_parquet("data.parquet")
print("\nParquet dtypes:\n", df_parquet.dtypes)
```

Observation on dtypes:
When reading the CSV, numeric columns may be inferred as float64 correctly, but sometimes strings or categorical-looking columns are read as object. In contrast, Parquet preserves the original data types exactly, so numeric columns remain float64 and the category column remains object. This happens because CSV is a text format without embedded type information, whereas Parquet is a typed binary format that stores schema metadata.

```python
import time
start_csv = time.time()
df_csv = pd.read_csv("data.csv")
end_csv = time.time()
csv_time = end_csv - start_csv
start_parquet = time.time()
df_parquet = pd.read_parquet("data.parquet")
end_parquet = time.time()
parquet_time = end_parquet - start_parquet

print(f"CSV read time: {csv_time:.4f} seconds")
print(f"Parquet read time: {parquet_time:.4f} seconds")
```

```python
start_col = time.time()
category_only = pd.read_parquet("data.parquet", columns=["category"])
end_col = time.time()
col_time = end_col - start_col

print(f"Parquet read (category only) time: {col_time:.4f} seconds")
```

Parquet is a columnar storage format, meaning each column is stored separately on disk. This allows reading only the needed columns (e.g., category) without loading all other columns into memory.
In contrast, CSV is a row-based text format, so reading any part of the file generally requires parsing all rows and columns. This makes selective column reads from CSV inefficient or impossible without additional tools.

Finding books by author and calculating average price is easiest with the flat or normalized tables, while the document format requires looping through nested editions. If books are usually accessed as standalone records, the document-oriented format is most convenient.
